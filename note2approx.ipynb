  <h2> Lecture-7 Taylor Approximations </h2>
 

```python
import numpy as np                      # Python library for efficient numerical calculations
import matplotlib.pyplot as plt         # Plot library
```

```python
x = np.arange(-3, 3, 0.2)             # defining the x range and step size
def linapprox(a):                        # Linear approximation to sin(x) near x=a                   
    return np.sin(a) + (x-a)*np.cos(a)
plt.plot(x, linapprox(0), 'b', x, np.sin(x), 'g')       
plt.show() 
```

The blue line is the tangent line to the curve, $y= \sin{x}$ at $x=0$. Note that the line passes through the point $(x, y) = (0,0)$ and has slope $\frac{dy}{dx}|_{x=0} = \cos{x}|_{x=0} = 1$. Hence the line is $y_l = x$ which is consistent with the small angle approximation $\sin{x} \approx x$ that you are familier with.

```python
x = np.arange(0, 1.5, 0.2)             # defining the x range and step size
def linapprox(a):                                           
    return np.sin(a) + (x-a)*np.cos(a)
plt.plot(x, linapprox(np.pi/4), 'b', x, np.sin(x), 'g')     # plotting linear approx to sin(x) at pi/4
plt.show() 
```

The blue line is the tangent line to the curve, $y= \sin{x}$ at $x=\pi/4$. Note that the line passes through the point $(x, y) = (\frac{\pi}{4}, \frac{1}{\sqrt{2}})$ and has slope $\frac{dy}{dx}|_{x=\pi/4} = \cos{x}|_{x=\pi/4} = \frac{1}{\sqrt{2}}$. Hence the line is $y_l = \frac{1}{\sqrt{2}} + \frac{1}{\sqrt{2}} (x - \pi/4)$. Note that both $y$ and $y_l$ have the same values and same slopes at the point $x = \frac{\pi}{4}$

```python
x = np.arange(-3, 3, 0.2)             # defining the x range and step size

def quadapprox(a):                                          
    return np.sin(a) + (x-a)*np.cos(a) + (x-a)**2/2*(-np.sin(a))
plt.plot(x, quadapprox(0), 'b', x, np.sin(x), 'g')          
plt.show() 
```

The blue line is the quadratic approximation to the curve, $y= \sin{x}$ at $x=0$. Note that the quadratic approximation is the same as the linear approximation because $\frac{d^2}{dx^2} \sin{x}$ at $x= 0$ is zero.

```python
x = np.arange(-.5, 1.8, 0.2)             # defining the x range and step size

def quadapprox(a):                                          
    return np.sin(a) + (x-a)*np.cos(a) + (x-a)**2/2*(-np.sin(a))
plt.plot(x, linapprox(np.pi/4), 'r', x, quadapprox(np.pi/4), 'b', x, np.sin(x), 'g')          
plt.show() 
```

The red line is the tangent line to green the curve, $y= \sin{x}$ at $x=\pi/4$, the blue curve is the quadratic approximation.

```python
x = np.arange(-.5, 1.8, 0.2)             # defining the x range and step size

def cubprox(a):                                          
    return np.sin(a) + (x-a)*np.cos(a) + (x-a)**2/2*(-np.sin(a))+(x-a)**3/6*(-np.cos(a))
plt.plot(x, cubprox(np.pi/4), 'r', x, quadapprox(np.pi/4), 'b', x, np.sin(x), 'g')          
plt.show() 
```

The red line is the cubic approximation to green the curve, $y= \sin{x}$ at $x=\pi/4$, the blue curve is the quadratic approximation.
